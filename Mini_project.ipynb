```python
# replace spaces and get lowercase
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

##### 2. Let's see the top-five countries with successful bookings

```python
df.query('is_canceled == 0').value_counts('country')
```

```python
df.query('is_canceled == 0').groupby('country').agg({'is_canceled': 'count'}).sort_values('is_canceled', ascending = False)
```

##### 3. Let's see how often two types of hotels are booked in average

```python
df[['hotel', 'stays_total_nights']]
```

```python
df.groupby('hotel', as_index = False).agg({'stays_total_nights': 'mean'}).round(2)
```

##### 4. Let's see overbooked rooms in the data

```python
df.query('reserved_room_type != assigned_room_type').shape
```

##### 5. Booking statistic per month in 2016 and 2017

```python
df.query('is_canceled == 0 & arrival_date_year == 2016')  \
  .groupby('arrival_date_month', as_index = False)  \
  .agg({'is_canceled': 'count'})  \
  .sort_values('is_canceled', ascending = False)
```

```python
df.query('is_canceled == 0 & arrival_date_year == 2017')  \
  .groupby('arrival_date_month', as_index = False)  \
  .agg({'is_canceled': 'count'})  \
  .sort_values('is_canceled', ascending = False)
```

```python
df.query('is_canceled == 1 & arrival_date_year == 2017 & hotel == "City Hotel"' )  \
  .groupby('arrival_date_month', as_index = False)  \
  .agg({'is_canceled': 'count'})  \
  .sort_values('is_canceled', ascending = False)
```

##### 6. Just analyze a couple of statistic indicators

```python
df.adults.mean()
```

```python
df.children.mean()
```

```python
df.babies.mean()
```

```python
print(max(df.adults.mean(), df.children.mean(), df.babies.mean()))
```

```python
df['total_kids'] = df['children'] + df['babies']
```

```python
df.groupby('hotel', as_index = False)  \
  .agg({'total_kids': 'mean'})  \
  .sort_values('total_kids', ascending = False).round(2)
```

```python
df
```

##### 7. Calculation of cancellations per hotel type per year

```python
df.query('is_canceled == 1 & hotel == "City Hotel"') \
  .groupby('arrival_date_year')['arrival_date_month'].value_counts()
```

##### 8. Calculation of churn rate

```python
df
```

```python
# cancelled bookings w/children / all bookings w/children
round(100 * df.query('is_canceled == 1 & total_kids != 0').shape[0] / df.query('total_kids != 0').shape[0], 2)
```

```python
# cancelled bookings without children / all bookings without children
round(100 * df.query('is_canceled == 1 & total_kids == 0').shape[0] / df.query('total_kids == 0').shape[0], 2)
```
